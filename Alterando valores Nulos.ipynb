```python
#Importando o Pandas
import pandas as pd

```

```python
#Carregando o arquivo na variável arquivo 
arquivo = "C:\\Users\\miche\\Desktop\\Codigos Python\\Datasets\\Alunos.txt"
#Utilizando o método read_table do Pandas e carregando os dados que estavam em modo tabular com o separador \t na variável 
#alunos
alunos = pd.read_table(arquivo)
#Lendo a variável alunos.
alunos
```

```python
#Verificando quais colunas no dataframe possuem valores nulos
alunos.isnull().sum()
```

```python
#Utilizando a função fillna ra alterar os NAN para 0 na coluna Nota_2_semestre e com inplace=true para alterar o dataframe 
#original.

alunos["Nota_2_semestre"].fillna(0, inplace=True)

#Utilizando a função fillna ra alterar os NAN para 'Sem Nome' na coluna Aluno e com inplace=true para alterar o dataframe 
#original.

alunos["Aluno"].fillna('Sem Nome', inplace=True)

#Verificando o dataframe alterado.
alunos
```
