```python
#Importando o Pandas
import pandas as pd
```

```python
#Carregando o arquivo na variável arquivo 
arquivo = "C:\\Users\\miche\\Desktop\\Codigos Python\\Datasets\\Alunos.txt"
#Utilizando o método read_table do Pandas e carregando os dados que estavam em modo tabular com o separador \t na variável 
#alunos
alunos = pd.read_table(arquivo)
#Lendo a variável alunos.
alunos
```

```python
#Filtrando o dataframe para trazer os alunos com nota maior ou igual a 5 no campo 'Nota_2_semestre'
alunos[alunos.Nota_2_semestre >= 5]
```

```python
#Aplicando multiplos filtros com operador E
#Trazer os alunos com nota maior ou igual a 5 no campo 'Nota_2_semestre' e nota maior ou igual a 6 no campo 'Nota_3_semestre'
alunos[(alunos.Nota_2_semestre >= 5) & (alunos.Nota_3_semestre >= 6)]
```

```python
#Aplicando multiplos filtros com o operador OU
#Trazer os alunos com nota maior ou igual a 5 no campo 'Nota_2_semestre' ou nota maior ou igual a 6 no campo 'Nota_3_semestre'
alunos[(alunos.Nota_2_semestre >= 5) | (alunos.Nota_3_semestre >= 6)]
```

```python
#Utilizando ISIN para filtra de acordo com uma lista. No caso abaixo trouxe uma lista com os nomes 'Rodolfo' e 'Paul'
alunos[alunos.Aluno.isin(['Rodolfo','Paul'])]
```
